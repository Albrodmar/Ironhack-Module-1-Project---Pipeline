<h2>Will get the GDP by countries from Wikipedia

```python
import requests
import pandas as pd
import numpy as np
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
```

```python
df = pd.read_html(url)
df[2] #there are several tables but just need this one
```

<h3>Will do so wrangling to put it nice

```python
gdp = df[2]
```

<h3>Renaming columns

```python
gdp.rename(columns = {0:'Rank', 1:'Country', 2:'GDP in Billions USD'}, inplace = True) #still values in millions
```

```python
gdp
```

<h3>There are various tables concatenated need to separate them. Last country of every serie is "Tuvalu"

```python
gdp[gdp['Country'].str.contains('Tuvalu')]
```

```python
gdp_new = gdp.iloc[198:388]
```

```python
gdp_new
```

```python
gdp_new = gdp_new.reset_index()
```

```python
gdp_new
```

```python
gdp_new = gdp_new.drop(columns = ['Rank','index'])
```

```python
gdp_new
```

<h3>Replacing at the Country/Territory column all the data in brakets with regex

```python
gdp_new['Country'] = gdp_new['Country'].str.replace(r'\[.+','')
```

<h3>The column GDP has values from 2017 and has added (2017) to the value 

```python
gdp_new['GDP in Billions USD'] = gdp_new['GDP in Billions USD'].str.replace(r'\ \(2017\)','')
gdp_new['GDP in Billions USD'] = gdp_new['GDP in Billions USD'].str.replace(r'\[.+','')
```

<h4>After replacing (2017) the values of GDP with 'comas' keep the coma and need to be erased

```python
gdp_new['GDP in Billions USD'] = gdp_new['GDP in Billions USD'].str.replace(',','')
```

<h4>Converting values of GDP fron object into integers and change the amounto from million to billions 

```python
#gdp_new['GDP in Billions USD'].values
#gdp['GDP in Billions USD'] = gdp['GDP in Billions USD'].astype(int)
gdp_new['GDP in Billions USD'] = gdp_new['GDP in Billions USD'].astype(int)
```

```python
gdp_new.dtypes
```

```python
gdp_new
```

```python
#converting from millions to billions
gdp_new['GDP in Billions USD'] = gdp_new['GDP in Billions USD']/1000
```
